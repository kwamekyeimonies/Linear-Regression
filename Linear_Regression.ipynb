```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
```

```python
read_data=pandas.read_csv('cost_revenue_clean.csv')
```

```python
read_data.describe()
```

```python
X=DataFrame(read_data,columns=["production_budget_usd"])
Y=DataFrame(read_data,columns=["worldwide_gross_usd"])
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X,Y,alpha=0.3)
plt.title("Movie Cost Vs Revenue")
plt.xlabel("Production Budget")
plt.ylabel("Worldwide Gross USD")
plt.ylim(0,3000000000)
plt.xlim(0,3000000000)
plt.show()
```
